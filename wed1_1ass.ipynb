```python
import sys

# Simulate input for demonstration in Colab, as sys.stdin.read() often returns empty
# You can replace this with actual input if running in an environment where stdin is populated.
input_data = """
1
4
4 20
1 10
1 40
1 30
"""
data = input_data.strip().split()

# Original code logic continues below
index = 0
T = int(data[index])
index += 1

results = []

for _ in range(T):
    N = int(data[index])
    index += 1

    jobs = []
    max_deadline = 0

    for _ in range(N):
        d = int(data[index])
        p = int(data[index + 1])
        index += 2
        jobs.append((d, p))
        max_deadline = max(max_deadline, d)

    # Sort jobs by profit descending
    jobs.sort(key=lambda x: x[1], reverse=True)

    slots = [-1] * (max_deadline + 1)

    count = 0
    total_profit = 0

    for d, p in jobs:
        for slot in range(d, 0, -1):
            if slots[slot] == -1:
                slots[slot] = p
                count += 1
                total_profit += p
                break

    results.append(f"{count} {total_profit}")

print("\n".join(results))
```

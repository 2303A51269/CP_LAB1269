```python
import bisect

def closest_subset_sum(arr, S):
    n = len(arr)
    mid = n // 2

    left = arr[:mid]
    right = arr[mid:]

    # Generate subset sums for left half
    left_sums = []
    for mask in range(1 << len(left)):
        total = 0
        for i in range(len(left)):
            if mask & (1 << i):
                total += left[i]
        left_sums.append(total)

    # Generate subset sums for right half
    right_sums = []
    for mask in range(1 << len(right)):
        total = 0
        for i in range(len(right)):
            if mask & (1 << i):
                total += right[i]
        right_sums.append(total)

    right_sums.sort()

    min_diff = float('inf')

    # Find minimum absolute difference
    for l_sum in left_sums:
        target = S - l_sum
        idx = bisect.bisect_left(right_sums, target)

        # Check closest candidates
        if idx < len(right_sums):
            min_diff = min(min_diff, abs(S - (l_sum + right_sums[idx])))

        if idx > 0:
            min_diff = min(min_diff, abs(S - (l_sum + right_sums[idx - 1])))

    return min_diff


# -------- Main Program --------
T = int(input().strip())

for _ in range(T):
    N, S = map(int, input().split())
    arr = list(map(int, input().split()))

    print(closest_subset_sum(arr, S))
```

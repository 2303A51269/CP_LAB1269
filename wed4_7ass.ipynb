```python
class FenwickTree:
    def __init__(self, n):
        self.n = n
        self.tree = [0] * (n + 1)

    # Add value to index i
    def update(self, i, delta):
        while i <= self.n:
            self.tree[i] += delta
            i += i & -i

    # Prefix sum from 1 to i
    def query(self, i):
        total = 0
        while i > 0:
            total += self.tree[i]
            i -= i & -i
        return total


# -------- Example Usage (Step Count Problem) --------

# Number of days
N = 7

# Daily step counts (1-based indexing)
steps = [4500, 6000, 5200, 7000, 4800, 6500, 5000]

# Create Fenwick Tree
ft = FenwickTree(N)

# Build tree
for i in range(N):
    ft.update(i + 1, steps[i])

# Query 1: total steps till Day 5
print("Total steps till Day 5 =", ft.query(5))

# Update Day 3 from 5200 â†’ 5800
old_value = steps[2]
new_value = 5800
delta = new_value - old_value
steps[2] = new_value
ft.update(3, delta)

# Query 2: total steps till Day 5 after update
print("After update, total steps till Day 5 =", ft.query(5))
```
